Central limit theorem is the idea that any distribution sampled over a long period of time will distribute normally. 

**Incorrect, if the i.i.d. random variables are suitably averaged in the way that a MC estimator does so, the distribution will tend towards normality as n increases.**

More specifically, we can describe the CDF of the normal function to be *equal* to the empirical CDF of the function as n tends towards infinity

### Exercise 40

All we need to do, is plot the empirical CDF alongside the standard normal CDF "Sigmoidical" function



```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
alpha = 2
beta = 2
mu = 0
std = 1
sample_size=1_000_000
B = np.random.beta(a=alpha,b=beta,size=sample_size) 
N = np.random.normal(mu,std,size=sample_size)
# parameters are alpha and beta

B=(B*3) - 4
```

```python
def EmpiricalCDF(X,x): 
    # Count how many values are "less than the target x"
    # Then divides by the number of target x.
    counter=0
    for i in X:
        if i < x:
            counter+=1
    return counter / len(X)
```

```python
#Generate points for plot
#We should have the x values we pass into function
#We should then have the resulting value for the EmpiricalCDF

array1_Beta = []
array1_Normal = []
array2 = np.arange(-3,3,step=0.05)

for i in array2:
    array1_Normal.append(EmpiricalCDF(N,i))
    array1_Beta.append(EmpiricalCDF(B,i))
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(array2,array1_Normal);
ax.plot(array2,array1_Beta,color='red');
```

The scales are a little wrong, but as you can see, the "normality" appears in the CDF of the beta function as we summate with larger values of steps in the `array2` range.
