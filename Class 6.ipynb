```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dice = [1,2,3,4,5,6]
sample_size = 500
D = np.random.choice(dice,size=sample_size,replace=True)
k = np.arange(1,sample_size+1)
Xbar = np.cumsum(D) / k
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(k,Xbar);
ax.set_ylim(int(min(Xbar))),int(max(Xbar))
ax.hlines(3.5,xmin=0,xmax=500,color='red');
```

Computing the variance of the sample mean is relative to the sample mean we calculate in the monte carlos method. Obviously the law of strong numbers states that 
$l  i  m_{n\to\infty}$ The value of $\bar{X}$ will reach closer and closer to $\mu$ of X.

This also means the variance will become closer and closer to the true variance of the population $X$.

Therefore to calulate a "running" variance, all we have todo is calculate

$$Var(\bar{X}) = E[(\bar{X})^2] - E[\bar{X}]^2$$

We know for variance with $Var(cX) = c^2Var(X)$

Hence we can just calculate the variance of the running mean by aproximating and filling in the values:

$$Var(\bar{X})$$

$$Var(\bar{X}) = Var(\frac{1}{n} * \sum X_i )$$

$$Var(\frac{1}{n} * \sum X_i ) = \frac{1}{n^2} * Var(X)$$

## Exercise 29

```python
def h(x):
    return np.cos(np.exp(3*x))

```

```python
sample_size=1_000_000
U = np.random.random(size=sample_size)
U = h(U)

U = np.cumsum(U)
X = np.arange(1,sample_size+1)
U = U / X

```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(X[50000:],U[50000:]);

ax.hlines(U.mean(),xmin=0,xmax=sample_size,color='red');
```
