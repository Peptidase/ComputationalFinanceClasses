### Exercise 9

$$
P(X \leqslant x) = F(x)
$$

We know the CDF for a r.v. is 

$P(RandomValue \leqslant value)$

We also know that

$ Y = \mu + \sigma * X $

So that means we can substitute:

$P(Y \leqslant y) = P(\mu + \sigma * X \leqslant y)$

And rearrange this into the following:

$ P(Y \leqslant y) = P(X \leqslant \frac{y-\mu}{\sigma})$

$ \therefore P(X \leqslant \frac{y-\mu}{\sigma}) = F(\frac{y-\mu}{\sigma})$

```python
sample_size = 1000
mu = 2
```

```python
U = np.random.uniform(-1,1,size=sample_size) #DUnif({-1,1}})
```

```python
x = np.linspace(0,1,sample_size)
X =stats.expon.pdf(x,scale=1/mu)
```

```python
DoubleEpon = X * U
```

```python
fig,ax = plt.subplots(4,figsize=(10,25))
ax[0].hist(DoubleEpon, bins=25);
ax[0].set_title("Doubly Exponentially distributed Random Variable")

ax[1].hist(X, bins=25,color='red');
ax[1].set_title("Exponentially distributed random variable")

ax[2].hist(U, bins=25, color='green');
ax[2].set_title("Uniformly distributed Random Variable");

ax[3].hist(DoubleEpon, bins=25);
ax[3].hist(X, bins=25,color='red');
ax[3].hist(U, bins=25, color='green');
ax[3].legend(['Doubly exponential','Uniformly distributed','Eponentially distributed'])
```
